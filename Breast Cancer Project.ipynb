```python
#creating the model to detect the cancer cell
#split the data set into independent (X) and dependent (Y) data sets
#will contain all columns after diagnosis, except the column at (index 30)
#were changed to arrays because of the parameters that will be taken in for the model

X = Breast_Cancer.iloc[:,2:31].values
Y = Breast_Cancer.iloc[:,1].values
```

```python
#dependent data set (i.e., X = will help to determine) if patient has cancer or not
#independent data set (i.e., Y = diagnosis) will tell the features that can detect if patient has cancer or not
```

```python
#need to split the data set into 75% training and 25% testing
#will train the model on 75% of the data and test the model on 25% of the data

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#25% of the data is X_test and Y_test correlates to the X_test data set
#75% of the data is X_train and Y_train correlates to the X_train data set
```

```python
#need to sclae the data to bring all the features to the same level of magnitute (this is called Feature Scaling)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#create a function for the models

def models(X_train, Y_train):
    
    #logistic regression model
    
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train, Y_train)
    
    #Decision Tree
    
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
    tree.fit(X_train, Y_train)
    
    #Random Forest Classifier
    
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    forest.fit(X_train, Y_train)
    
    #Print the models accuracy on the training data
    
    print('[0]Logistic Regression Training Accuracy:', log.score(X_train, Y_train))
    print('[1]Decision Tree Classifier Training Accuracy:', tree.score(X_train, Y_train))
    print('[2]Random Forest Classifier Training Accuracy:', forest.score(X_train, Y_train))
    
    return log, tree, forest
```

```python
#getting all of the models

model = models(X_train, Y_train)
```

```python
#test the model accuracy on the testing data using confusion matrix 

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, model[0].predict(X_test))

print(cm)
```

```python
#true postive value = 86; true negative value = 50; false positive value = 4; false negative value = 3
#to get the accuracy of the model on the testing data, need to calculate the values
#true positive = TP' true neagtive = TN; false postive = FP' false neagtive = FN

TP = cm[0][0]
TN = cm [1][1]
FN = cm [1][0]
FP = cm [0][1]

print('Testing Accuracy = ', (TP + TN)/ (TP + TN + FN + FP))
```

```python
#the logistic regression model's accuracy on the testing data is ~ 95.1%
#get accuracy for all models; put in a loop

from sklearn.metrics import confusion_matrix

for i in range( len(model) ):
    print('Model', i)
    cm = confusion_matrix(Y_test, model[i].predict(X_test))
    
    TP = cm[0][0]
    TN = cm [1][1]
    FN = cm [1][0]
    FP = cm [0][1]
    
    print(cm)
    print('Testing Accuracy = ', (TP + TN)/ (TP + TN + FN + FP))
    print()
    
    #model 0 = logisitc regression
    #model 1 = decision tree
    #model 2 = random forest 
```

```python
#show another, easier way to get metics of the models

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

print( classification_report(Y_test, model[0].predict(X_test)))
print( accuracy_score(Y_test, model[0].predict(X_test)))
```

```python
#show another way to get metics of the models using a loop

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range( len(model) ):
    print('Model', i)
    print( classification_report(Y_test, model[i].predict(X_test)))
    print( accuracy_score(Y_test, model[i].predict(X_test)))
    print()
```
